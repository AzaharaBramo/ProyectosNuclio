### <span style="color:#3498db"> **1.2 QUERY COMBINADA**

```python
# ========== QUERY COMBINADA - TODOS LOS FILTROS ==========

query_final = """
SELECT 
    a.CODIGO_ALOJAMIENTO,
    a.NOMBRE,
    a.TERRAZA,
    a.AIREC_ACOND, 
    a.PISCINA,
    a.SALON,
    a.CANT_BANYOS,
    a.SUPERFICIE,
    u.PAIS,
    u.CIUDAD,
    u.DIST_METRO,
    u.DIST_CENTRO,
    p.PRECIO_ALQUILER,
    p.PORCENTAJE_RESERVA,
    pt.PUNTOS,
    pt.AGENCIA,
    pt.PUNTOS_AGENCIA
FROM alojamiento a
JOIN ubicacion u ON a.CODIGO_ALOJAMIENTO = u.CODIGO_ALOJAMIENTO
JOIN precio p ON a.CODIGO_ALOJAMIENTO = p.CODIGO_ALOJAMIENTO
JOIN puntuacion pt ON a.CODIGO_ALOJAMIENTO = pt.CODIGO_ALOJAMIENTO
WHERE a.TERRAZA = 'SI'
  AND a.AIREC_ACOND = 'SI'
  AND a.PISCINA = 'NO'
  AND a.SALON = 'SI'
  AND a.CANT_BANYOS > 1
  AND a.SUPERFICIE >= 80
  AND u.DIST_METRO > 1
  AND u.DIST_CENTRO > 2
  AND p.PRECIO_ALQUILER BETWEEN 1500 AND 2000
  AND p.PORCENTAJE_RESERVA = 25
  AND pt.PUNTOS > 4.5
  AND pt.PUNTOS_AGENCIA > 4
"""

pd.read_sql(query_final, conn)
```

### <span style="color:#3498db"> **1.3 RECOMENDACI√ìN**

 <span style="color:#3498db"> Como hab√©is visto, tenemos 4 propiedades que cumplen con todas las exigencias de nuestro cliente, ahora bien, desde tu punto de vista, haciendo un an√°lisis de los datos del cliente y de todas sus exigencias ¬øCu√°l de estos 4 alojamientos le recomendar√≠as y por qu√©?

```python
# ========== QUERY COMBINADA - INCLUYENDO HABITACIONES ==========

query_final = """
SELECT 
    a.CODIGO_ALOJAMIENTO,
    a.NOMBRE,
    a.CANT_HAB,
    a.CANT_BANYOS,
    a.SUPERFICIE,
    a.TERRAZA,
    a.AIREC_ACOND, 
    a.PISCINA,
    a.SALON,
    u.PAIS,
    u.CIUDAD,
    u.DIST_METRO,
    u.DIST_CENTRO,
    p.PRECIO_ALQUILER,
    p.PORCENTAJE_RESERVA,
    pt.PUNTOS,
    pt.AGENCIA,
    pt.PUNTOS_AGENCIA
FROM alojamiento a
JOIN ubicacion u ON a.CODIGO_ALOJAMIENTO = u.CODIGO_ALOJAMIENTO
JOIN precio p ON a.CODIGO_ALOJAMIENTO = p.CODIGO_ALOJAMIENTO
JOIN puntuacion pt ON a.CODIGO_ALOJAMIENTO = pt.CODIGO_ALOJAMIENTO
WHERE a.TERRAZA = 'SI'
  AND a.AIREC_ACOND = 'SI'
  AND a.PISCINA = 'NO'
  AND a.SALON = 'SI'
  AND a.CANT_BANYOS > 1
  AND a.SUPERFICIE >= 80
  AND u.DIST_METRO > 1
  AND u.DIST_CENTRO > 2
  AND p.PRECIO_ALQUILER BETWEEN 1500 AND 2000
  AND p.PORCENTAJE_RESERVA = 25
  AND pt.PUNTOS > 4.5
  AND pt.PUNTOS_AGENCIA > 4
ORDER BY a.CANT_HAB DESC, a.SUPERFICIE DESC
"""

pd.read_sql(query_final, conn)
```

### <span style="color:#3498db">2.9 QUERY CON OR  Y AND ("PISO" O PISCINA Y A/C, <150m o >1 wc)

```python
# Query 9: Tres condiciones separadas
query9 = """SELECT CODIGO_ALOJAMIENTO, NOMBRE 
FROM alojamiento 
WHERE NOMBRE LIKE '%Piso%' 
   OR (PISCINA = 'SI' AND AIREC_ACOND = 'SI' AND SUPERFICIE < 150)
   OR (SUPERFICIE >= 150 AND CANT_BANYOS > 1)"""

pd.read_sql(query9, conn)
```

```python
# Query 9: Alojamientos que cumplan cualquiera de las 4 
query9 = """SELECT CODIGO_ALOJAMIENTO, NOMBRE 
FROM alojamiento 
WHERE NOMBRE LIKE '%Piso%' 
   OR (PISCINA = 'SI' AND AIREC_ACOND = 'SI' AND SUPERFICIE < 150) 
   OR CANT_BANYOS > 1"""

pd.read_sql(query9, conn)
```

Este enunciado tiene 4 condiciones. Pero cada dos condiciones se cancela con un OR, es decir finalmente tiene que cumplir una condicio√≥n de cada par. Devolver√°:
- Si es PISO <150m 
- Si es PISO > 1 wc. 
- Si tiene piscina+AC, <150m
- Si tiene piscina+AC, >1wc. 

```python
# Query 9b: Dos grupos combinados correctamente a mi parecer esta es la query que har√≠a para el resultado que intuyo busca en el enunciado
query9b = """SELECT CODIGO_ALOJAMIENTO, NOMBRE 
FROM alojamiento 
WHERE (NOMBRE LIKE '%Piso%' AND (SUPERFICIE < 150 OR CANT_BANYOS > 1))
   OR (PISCINA = 'SI' AND AIREC_ACOND = 'SI' AND (SUPERFICIE < 150 OR CANT_BANYOS > 1))"""

pd.read_sql(query9b, conn)
```

### <span style="color:#3498db">2.10 NUEVA COLUMNA 

```python
# Query 10: Mostrar precio normal y precio con s√≠mbolo ‚Ç¨
query10 = """SELECT 
    a.CODIGO_ALOJAMIENTO,
    a.NOMBRE,
    p.PRECIO_ALQUILER,
    CONCAT(p.PRECIO_ALQUILER, '‚Ç¨') as PRECIO_MODIF
FROM alojamiento a
JOIN precio p ON a.CODIGO_ALOJAMIENTO = p.CODIGO_ALOJAMIENTO"""

pd.read_sql(query10, conn)
```

### <span style="color:#3498db">2.11 CATEGORIA PRECIO

```python
# Query 11: Columna CATEGORIA_PRECIO con categor√≠as seg√∫n rangos
query11 = """SELECT 
    CODIGO_ALOJAMIENTO,
    PRECIO_ALQUILER,
    RESERVA,
    PORCENTAJE_RESERVA,
    CASE 
        WHEN PRECIO_ALQUILER BETWEEN 780 AND 1000 THEN 'Bajo'
        WHEN PRECIO_ALQUILER BETWEEN 1001 AND 1700 THEN 'Medio' 
        WHEN PRECIO_ALQUILER > 1700 THEN 'Alto'
        ELSE 'Sin categor√≠a'
    END AS CATEGORIA_PRECIO
FROM precio
ORDER BY PRECIO_ALQUILER"""

pd.read_sql(query11, conn)
```
