```python
# Filtrar viviendas enteras en el Centro con precio por huésped <= 20€
baratos = df_centro[(df_centro['room_type'] == 'entire_home_apt') & 
                    (df_centro['price_per_guest'] <= 20)]

# Mostrar algunos ejemplos
baratos[['id', 'price', 'accommodates', 'price_per_guest']].sort_values(by='price_per_guest').head(50)
```

* El aire acondicionado es el servicio menos presente (79 % en baratos vs. 83 % en caros), y puede justificar una subida de precio.
* Aquí sumaremos la necesidad de añadir aire acondicionado en aquellos que no tengan, ya que en ambos grupos, baratos o caros está por debajo de 0,9.
* Tambien valorar el reconvertir esos apartametnos enteros de número de huéspedes elevados (>6) a máximo 6 huéspedes que se alquilan más. 

```python
# 1. Filtrar apartamentos enteros en el Centro
df_centro_entero = df_price_filtered[
    (df_price_filtered['neighbourhood_group'] == 'Centro') &
    (df_price_filtered['room_type'] == 'entire_home_apt')
]

# 2. Agrupar por número de huéspedes permitidos
df_ocupacion = df_centro_entero.groupby('accommodates')['reviews_per_month'].mean()

# 3. Visualizar
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
df_ocupacion.plot(kind='bar')
plt.title('Media de reviews por mes según capacidad (Centro, aptos enteros)')
plt.xlabel('Número de huéspedes permitidos')
plt.ylabel('Reviews por mes (media)')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
# 1. Filtrar solo apartamentos enteros en el Centro
df_centro_entero = df_price_filtered[
    (df_price_filtered['neighbourhood_group'] == 'Centro') &
    (df_price_filtered['room_type'] == 'entire_home_apt')
].copy()

# 2. Crear una nueva columna que divide las reviews por mes entre el número de huéspedes permitidos.
# Esto estima cuán "rentable" es cada cama (más o menos), basándose en la cantidad de reviews por huesped permitido.
df_centro_entero['reviews_per_month_per_guest'] = (
    df_centro_entero['reviews_per_month'] / df_centro_entero['accommodates']
)

# 3. Agrupar por capacidad y calculamos la media de ese ratio
df_ratio = df_centro_entero.groupby('accommodates')['reviews_per_month_per_guest'].mean()

# 4. Representar con un gráfico de barras
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
df_ratio.plot(kind='bar')
plt.title("Media de reviews por mes *por huésped* según capacidad (Centro, aptos enteros)")
plt.xlabel("Número de huéspedes permitidos")
plt.ylabel("Reviews por mes por huésped (media)")
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

* Con este gráfico vemos que a mayor número de huéspedes por apartamento completo disminuye la ratio de reviews. Con eso deduzco que la reserva es menor en esos apartamentos y saco un insight más. 
* LOS APARTAMENTOS DE GRAN CAPACIDAD SE ALQUILAN MENOS.

### 3.2.6 Hipotesis: las propiedades más caras son por el host


```python
# Definir los barrios top
top_barrios = ['Moncloa - Aravaca', 'Retiro', 'Chamberí']

# Lista de columnas del host
host_cols = [
    'host_response_rate',
    'host_is_superhost',
    'host_has_profile_pic',
    'host_identity_verified',
    'host_verifications_email',
    'host_verifications_phone',
    'host_verifications_reviews',
    'host_verifications_facebook',
    'host_verifications_government_id',
    'calculated_host_listings_count',
    'host_seniority'
]

# Separar DataFrames
df_top = df_price_filtered[df_price_filtered['neighbourhood_group'].isin(top_barrios)]
df_otros = df_price_filtered[~df_price_filtered['neighbourhood_group'].isin(top_barrios)]

# Calcular medias
host_top = df_top[host_cols].mean(numeric_only=True)
host_otros = df_otros[host_cols].mean(numeric_only=True)

# Comparar en un DataFrame
comparativa_host = pd.DataFrame({
    'top_barrios': host_top,
    'otros_barrios': host_otros
})
comparativa_host['diferencia'] = comparativa_host['top_barrios'] - comparativa_host['otros_barrios']

# Ordenar por diferencia absoluta
comparativa_host = comparativa_host.reindex(comparativa_host['diferencia'].abs().sort_values(ascending=False).index)

comparativa_host
```

#### Aunque las diferencias son muy pequeñas las más destacadas son:
* Mayor antigüedad (host_seniority): En media, los anfitriones de estos barrios llevan más tiempo activos en Airbnb (20 días más). Esto puede estar asociado a una mejor gestión o reputación consolidada.

* Más propiedades publicadas (calculated_host_listings_count): Tienen una media de 1,3 viviendas más por anfitrión, lo que podría indicar mayor profesionalización.

#### Conclusión: Aunque las diferencias son pequeñas, estos dos factores sugieren que parte del mayor precio puede deberse a hosts más experimentados y con mayor volumen de gestión.
